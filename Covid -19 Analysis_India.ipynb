```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data1=pd.read_csv('https://raw.githubusercontent.com/smithaprakash91/Covid-19-India/main/covid_19_india.csv')
data2=pd.read_csv('https://raw.githubusercontent.com/smithaprakash91/Covid-19-India/main/StatewiseTestingDetails.csv')
data3=pd.read_csv('https://raw.githubusercontent.com/smithaprakash91/Covid-19-India/main/covid_vaccine_statewise.csv')
```

```python
data1=data1.drop(['Sno','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
data1
```

```python
data1['Active']=data1['Confirmed']-data1['Cured']-data1['Cured']
data1
```

```python
data1['Date']=pd.to_datetime(data1['Date'])
data1
```

```python
fig=plt.figure(figsize=(15,8))
sns.lineplot(data=data1[data1['State/UnionTerritory'].isin(['Kerala','Tamil Nadu','Delhi','Maharashtra','Uttar Pradesh','Karnataka'])], x='Date', y='Active', hue='State/UnionTerritory')
plt.title('Active Cases by State')
```

```python
Total_cases=data1[data1['Date']==data1['Date'].max()].copy().fillna(0)
Total_cases.index=Total_cases['State/UnionTerritory']
Total_cases=Total_cases.drop(['State/UnionTerritory','Date'],axis=1)
Total_cases.head(10)
```

```python
top_10_states_deaths=Total_cases.sort_values('Deaths',ascending=False)[:10]
top_10_states_deaths
```

```python
top_10_states_deaths=Total_cases.sort_values('Deaths', ascending=False)[:10]
fig=plt.figure(figsize=(15,8))
plt.title('Top 10 States with highest death rate', size=20)
plt.xlabel('States')
plt.ylabel('Number of deaths')
ax=sns.barplot(data=top_10_states_deaths[:10].iloc[:10],y=top_10_states_deaths[:10].Deaths,x=top_10_states_deaths[:10].index, linewidth=2).set(xlabel='States', ylabel='Number of deaths')
plt.show()
```

```python
top_10_states_cured=Total_cases.sort_values('Cured', ascending=False)[:10]
fig=plt.figure(figsize=(15,8))
plt.title('Top 10 States with highest cured rate', size=20)
plt.xlabel('States')
plt.ylabel('Number of cured cases')
ax=sns.barplot(data=top_10_states_cured[:10].iloc[:10],y=top_10_states_cured[:10].Deaths,x=top_10_states_cured[:10].index, linewidth=2).set(xlabel='States', ylabel='Number of cured cases')
plt.show()
```

```python
sample_collected=data2[data2['Date']==data2['Date'].max()].copy().fillna(0)
sample_collected.index = sample_collected["State"]
sample_collected = sample_collected.drop(['Date'], axis=1)
sample_collected.head(10)
```

```python
top_10_states_screening=sample_collected.sort_values('TotalSamples', ascending=False)[:10]
fig=plt.figure(figsize=(15,8))
plt.title('Top 10 States Samples collected', size=20)
plt.xlabel('States')
plt.ylabel('Number of samples collected')
ax=sns.barplot(data=top_10_states_screening[:10].iloc[:10],y=top_10_states_screening[:10].TotalSamples,x=top_10_states_screening[:10].index, linewidth=2).set(xlabel='States', ylabel='Samples collected statewise')
plt.show()
```

```python
labels = 'Negative','Positive','TotalSamples'
sizes= [15,30,45]
explode=(0,0,0)

fig,ax=plt.subplots(figsize=(15,8))
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Total Number of samples', size=20)
```
